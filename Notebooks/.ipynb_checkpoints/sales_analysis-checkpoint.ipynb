```python
import pandas as pd
import numpy as np
```

```python
data = {
    "OrderID": [1, 2, 3, 4, 5],
    "Product": ["Pen", "Book", "Pen", "Bag", "Book"],
    "Category": ["Stationery", "Stationery", "Stationery", "Accessories", "Stationery"],
    "Quantity": [10, 5, 7, 2, 3],
    "Price": [10, 50, 10, 500, 50],
    "City": ["Chennai", "Madurai", "Chennai", "Trichy", "Madurai"]
}

df = pd.DataFrame(data)

df
```

```python
df["TotalSales"] = df["Quantity"] * df["Price"]

df
```

```python
df["TotalSales"].sum()
```

```python
df["OrderID"].count()
```

```python
df.groupby("Product")["TotalSales"].sum().sort_values(ascending=False)
```

```python
df.groupby("Category")["TotalSales"].sum()
```

```python
df.groupby("City")["TotalSales"].sum()
```

```python
df["TotalSales"].max()
```

```python
df["TotalSales"].mean()
```

```python
df["Quantity"].sum()
```

```python
df["Product"].nunique()
```

```python
summary = pd.DataFrame({
    "Total Sales": [df["TotalSales"].sum()],
    "Total Orders": [df["OrderID"].count()],
    "Total Quantity": [df["Quantity"].sum()],
    "Average Order Value": [df["TotalSales"].mean()]
})

summary
```
