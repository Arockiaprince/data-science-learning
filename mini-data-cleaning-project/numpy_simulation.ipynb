```python
import numpy as np
```

```python
rolls = np.random.randint(1, 7, 10000)

print(rolls)
```

```python
expected_value = np.mean(rolls)

print("Expected Value:", expected_value)
```

```python
variance = np.var(rolls)

print("Variance:", variance)
```

```python
for i in range(1, 7):
    probability = np.sum(rolls == i) / len(rolls)
    print(f"Probability of {i}: {probability}")
```

```python
import matplotlib.pyplot as plt

values = range(1, 7)
counts = [np.sum(rolls == i) for i in values]

plt.bar(values, counts)
plt.xlabel("Dice Value")
plt.ylabel("Frequency")
plt.title("Dice Roll Simulation")
plt.show()
```
