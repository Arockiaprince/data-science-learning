```python
# Grade Calculator

scores = []

num_subjects = int(input("Enter number of subjects: "))

for i in range(num_subjects):

    mark = int(input(f"Enter mark for subject {i+1}: "))

    # Validate marks
    while mark < 0 or mark > 100:
        print("Invalid mark. Enter between 0 and 100.")
        mark = int(input(f"Re-enter mark for subject {i+1}: "))

    scores.append(mark)

# Calculate statistics
total = sum(scores)
average = total / len(scores)
highest = max(scores)
lowest = min(scores)

# Grade calculation
if average >= 90:
    grade = "A"
elif average >= 75:
    grade = "B"
elif average >= 50:
    grade = "C"
else:
    grade = "Fail"

# Display result
print("\n----- Grade Summary -----")
print("Scores:", scores)
print("Total Marks:", total)
print("Average:", average)
print("Highest Mark:", highest)
print("Lowest Mark:", lowest)
print("Final Grade:", grade)
```
